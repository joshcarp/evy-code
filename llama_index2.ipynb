```python
from llama_index.readers.huggingface_fs import HuggingFaceFSReader
loader = HuggingFaceFSReader()
from llama_index.core import Document
documents = loader.load_data("datasets/joshcarp/evy-code")
from datasets import load_dataset
dataset = load_dataset("joshcarp/evy-code")
documents = []
for example in dataset["train"]:
    documents.append(Document(text=example["content"], metadata={"repo_url": example["repo_url"], "file_path": example["file_path"]}))
```

```python

# bge embedding model
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.core.embeddings import resolve_embed_model
from llama_index.llms.ollama import Ollama
from llama_index.readers.file import CSVReader
from llama_index.readers.huggingface_fs import HuggingFaceFSReader
Settings.embed_model = resolve_embed_model("local:BAAI/bge-small-en-v1.5")

# ollama
Settings.llm = Ollama(model="mistral", request_timeout=30.0)

index = VectorStoreIndex.from_documents(
    documents,
)
```

```python
query_engine = index.as_query_engine()
response = query_engine.query("""Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
// You may assume that each input would have exactly one solution, and you may not use the same element twice.
// You can return the answer in any order. write the code in evy for me""")
print(response)
```
